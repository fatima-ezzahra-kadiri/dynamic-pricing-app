```python
# Supposons que ton DataFrame s'appelle df
rows_all_null = df[df.isnull().all(axis=1)]

print(rows_all_null)
```

```python
# Supprimer uniquement les lignes où 'dynamic_price' est NaN
df = df.dropna(subset=['dynamic_price'])
 
print(df.shape)
```

```python
null_counts = df.isnull().sum()
print(null_counts)
```

```python
df.columns
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

 
columns_to_plot = [
i for i in df.columns
]

# Créer un boxplot pour chaque colonne
plt.figure(figsize=(12, 8))
for i, col in enumerate(columns_to_plot, 1):
    plt.subplot(2, 4, i)  # 2 lignes, 4 colonnes
    sns.boxplot(y=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
Q1 = df['dynamic_price'].quantile(0.25)
Q3 = df['dynamic_price'].quantile(0.75)
IQR = Q3 - Q1

# Définir les bornes
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrer les données
df = df[(df['dynamic_price'] >= lower_bound) & (df['dynamic_price'] <= upper_bound)]
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

 
columns_to_plot = [
i for i in df.columns
]

# Créer un boxplot pour chaque colonne
plt.figure(figsize=(12, 8))
for i, col in enumerate(columns_to_plot, 1):
    plt.subplot(2, 4, i)  # 2 lignes, 4 colonnes
    sns.boxplot(y=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
 
features = df.drop('dynamic_price', axis=1).columns
 

# Calcul de la matrice de corrélation
corr_matrix = df[features].corr()

# Affichage
plt.figure(figsize=(10, 7))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Matrice de corrélation des features")
plt.show()
```

les variables sont faiblement corrélé, donc pas contexte d'ACP (- L’ACP cherche à réduire la dimensionnalité en combinant des variables corrélées pour former des composantes principales. ) envisager d’autres méthodes de réduction de dimension (par exemple sélection de variables basée sur importance

```python
# Corrélation des features avec la target
target_corr = df[features].corrwith(df['dynamic_price'])

# Affichage sous forme de tableau
print(target_corr)

# Optionnel : visualisation en barplot
plt.figure(figsize=(10, 6))
sns.barplot(
    x=target_corr.index,
    y=target_corr.values,
    hue=target_corr.index,   # utiliser x comme hue
    palette="coolwarm",
    legend=False
)

plt.xticks(rotation=45)
plt.title("Corrélation des features avec la target (dynamic_price)")
plt.ylabel("Corrélation")
plt.show()
```

```python
median_value = df['demand_index'].median()
df['demand_index'] = df['demand_index'].fillna(median_value)
```

```python
df['demand_index'].unique()
```

```python
median_value = df['time_slot'].median()
df['time_slot'] = df['time_slot'].fillna(median_value)
```

```python
df.time_slot.unique()
```

```python
import numpy as np

arr = df['time_slot'].values

# Étape 1 : rendre les valeurs positives
arr = np.where(arr < 0, -arr, arr)

# Étape 2 : limiter à 23 si supérieur
arr_transformed = np.where(arr > 23, 23, arr)

# Remplacer dans le DataFrame
df['time_slot'] = arr_transformed
```

```python
df.time_slot.unique()
```

```python
df.day_of_week.unique()
```

```python
mode_value = df['day_of_week'].mode()[0]
df['day_of_week'] = df['day_of_week'].fillna(mode_value)
```

```python
df['day_of_week'].unique()
```

```python
df['day_of_week'].nunique()
```

```python
df['day_of_week'].value_counts()
```

puisque les occurences de day les plus répétes sont 6 et 0, dons on va prendre en compte la numérotation entre 0 et 6 et qui est bien de longeur 7

```python
df['day_of_week'] = df['day_of_week'].clip(upper=6)
```

```python
df['day_of_week'].value_counts()
```

```python
mean_value = df['competition_pressure'].mean()
df['competition_pressure'] = df['competition_pressure'].fillna(mean_value)
```

```python
median_value = df['operational_cost'].median()
df['operational_cost'] = df['operational_cost'].fillna(median_value)
```

```python
median_value = df['seasonality_index'].mean()
df['seasonality_index'] = df['seasonality_index'].fillna(median_value)
```

```python
median_value = df['marketing_intensity'].mean()
df['marketing_intensity'] = df['marketing_intensity'].fillna(median_value)
```
